in many forms, and involves producers, consumers and value-added uses of data.  Recently,
data has become the source of contention, especially as large companies continue to take control
of individual's data and use it in ways that do no benefit, and in some cases
are used in ways that were neither approved or known by, the the individual directly.

There are a number of interesting movements in data protection, such as the European Union's GDPR
(General Data Protection Regulation - [https://eugdpr.org](https://eugdpr.org)) that are trying to
hold companies account for how individual's data are collected, accessed, shared, distributed and protected.

access to controlling data in ways that would benefit the individual directly, should they
want to have it knowingly used and offered as a product to companies.  In scenarios like these,
the individual is then brought into the data transaction as a producer, and companies act as
consumers that add value to other consumers, etc.

development of decentralized and distributed data "ledgers" that may provide a more complete
platform to share data, trace it's lineage, but also validate and facilitate data transactions

2. assign a state code to the zip codes,


```python
df_zip = pd.read_csv("US.txt", header=None, sep='\t')
```

```python
zip_changed = data[~rows_with_dashes]
updated_zip_col = zip_changed[zip_changed.SPRADDR_ZIP_PR.map(str).map(len) == 5]
updated_zip_col = pd.DataFrame(updated_zip_col)
updated_zip_col
```

```python
#updated_zip_col['SPRADDR_ZIP_PR'] = updated_zip_col['SPRADDR_ZIP_PR'].astype(str)
```

```python
def getStateCode(z):
    try:
        z = df_zip[df_zip[1]==int(z)][4].item()
    except:
        pass
    return z
```

```python
updated_zip_col['SPRADDR_ZIP_PR'] = updated_zip_col['SPRADDR_ZIP_PR'].astype(int).apply(getStateCode)
```

```python
updated_zip_col['SPRADDR_ZIP_PR']
```

```python
updated_zip_col['SPRADDR_ZIP_PR'] = updated_zip_col['SPRADDR_ZIP_PR'].astype(str)
```

```python
updated_zip_col['SPRADDR_ZIP_PR'] = updated_zip_col['SPRADDR_ZIP_PR'][~updated_zip_col['SPRADDR_ZIP_PR'].str.isnumeric()]
```

```python
updated_zip_col = updated_zip_col.dropna()
updated_zip_col.head()
```

3. get an aggregate count of all enrollment numbers by state from 2014-2019.

```python
agg_count_all_enr = updated_zip_col.groupby('SPRADDR_ZIP_PR').count()
```

```python
agg_count_all_enr = agg_count_all_enr.reset_index()
agg_count_all_enr[['SPRADDR_ZIP_PR','ENROLLMENT']]
```
